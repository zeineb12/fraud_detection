```python
# Tokenize content based on words and ignores one-character words and punctuations
tokenizer = RegexpTokenizer(r'\b\w\w+\b')

# Here we need to drop the nan values we have talked about earlier or we'll get errors while tokenizing
data['subject_word_count'] = data['Subject'].map(
    lambda x: len(tokenizer.tokenize(x)))
data['content_word_count'] = data['content'].map(
    lambda x: len(tokenizer.tokenize(x)))
```

# 2.Network Analysis: <a class="anchor" id="network_analysis"></a>
We create a graph having as nodes as the name of employees with edges representing the connection between them. We will create an index between 0 and 1 that represents the strength of the connection between the nodes. This will alow us to see which people are interacting with who and therefore try to seperates the several departments and entities inside the firm.

## 2.1.Draw (part of) the Network: <a class="anchor" id="draw_network"></a>

```python
network_data = data[['From', 'To']]
network_data.head(3)
```

**We notice here that an email could be sent to multiple people. Splitting the 'To' column would look like this**

```python
test = network_data.head(3)
df = pd.DataFrame(test.To.str.split(',').tolist())
df
```

**We'll only use the first recipient for the graph to make things simpler cause sending a message to at least one person is the only guarantee we have.**

```python
df = pd.DataFrame(
    network_data.To.str.split(',', 1).tolist(),
    columns=['recipient', 'recipient2'],
    index=network_data.index)
network_data['To'] = df.copy()['recipient']
network_data.head()
```

```python
# We'll use a random sample from the entire dataset in order for this to run smoothly
network_data = network_data.sample(5000)

# Get top senders and receivers to label the graph
top_senders = network_data.groupby('From').count().sort_values(
    by='To', ascending=False).head(5).index.values
top_receivers = network_data.groupby('To').count().sort_values(
    by='From', ascending=False).head(5).index.values

# Get the graph
graph = nx.from_pandas_edgelist(network_data, 'From', 'To')
```

```python
labels = {}
color_map = []

for node in graph.nodes():
    if (node in top_senders) or (node in top_receivers):
        # Set the node name as the key and the label as its value
        labels[node] = node
        # Set color of the important node
        color_map.append('r')
    else:
        color_map.append((0.1, 0.5, 0.7, 0.6))

for key, value in labels.items():
    labels[key] = labels[key].split('@')[0]
```

```python
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(graph, k=0.2)

#set the argument 'with labels' to False so you have unlabeled graph
nx.draw_networkx(
    graph,
    pos,
    node_size=25,
    node_color=color_map,
    with_labels=False,
    edge_color='gray')
#Now only add labels to the nodes you require (the hubs in my case)
nx.draw_networkx_labels(
    graph,
    pos,
    labels=labels,
    font_size=9,
    font_weight='bold',
    font_color='black')

plt.show()
```

## 2.2. How close are employees to each other ? <a class="anchor" id="connectivity"></a>

The `degree_centrality` function of networkxx gives us an idea on how well each person is connected to the other

```python
centrality = nx.degree_centrality(graph)
name = []
degree_centrality = []

for key, value in centrality.items():
    name.append(key.split('@')[0])
    degree_centrality.append(value)

df_centrality = pd.DataFrame()
df_centrality['name'] = name
df_centrality['centrality'] = degree_centrality
df_centrality = df_centrality.sort_values(
    by='centrality', ascending=False).head(10)

# Plot data
plot_barh(
    (15, 8), df_centrality['name'], df_centrality['centrality'], (0.2, 0.4, 0.6, 0.6),
    'Degree of connectivity',
    'Who are the most connected pepole at ENRON ?')
```

We can see here that Kay Mann, Tana Jones , Vince Kaminski and Jeff Dasovich were the most connected people inside ENRON, they are the people who have the maximum number of contacts within the company. This can let us think that these people are mostly responsible of the fraud but we will see later that it is not the case.

Their high connectivity to employees could be due to the fact that these people work in the human ressources department and therefore communicate a lot with employees.

In this section we will convert our collection of e-mails to a matrix of token counts (Bag of Words). We will then do the weighting and normalization of words using TF-IDF techniques.

```python
messages_matrix = BOW_transformer.transform(X_train)

tfidf_transformer = TfidfTransformer().fit(messages_matrix)

messages_tfidf = tfidf_transformer.transform(messages_matrix)

test = tfidf_transformer.transform(BOW_transformer.transform(X_test))
```

## 3.3 Apply different machine learning algorithms :

### 3.3.1 Define ML algorithms

```python
random_state = 43

names = [
    "Naive Bayesian Multinomial",
    "k-Nearest Neighbors",
    "Support Vector Machine",
    "Linear SVM",
    "RBF SVM",
    "Extra Trees",
    "Random Forest",
    "Extra Forest",
    "AdaBoost",
    "Logistic Regression",
    "SGD Classifier",
]

algorithms = [
    MultinomialNB(),
    KNeighborsClassifier(n_neighbors=3),
    SVC(random_state=random_state),
    SVC(kernel="linear", random_state=random_state),
    SVC(kernel="rbf", random_state=random_state),
    ExtraTreesClassifier(random_state=random_state),
    RandomForestClassifier(random_state=random_state),
    GradientBoostingClassifier(random_state=random_state),
    AdaBoostClassifier(
        DecisionTreeClassifier(random_state=random_state),
        n_estimators=10,
        learning_rate=0.1,
        random_state=random_state),
    LogisticRegression(random_state=random_state),
    SGDClassifier(max_iter=1000),
]
```

### 3.3.2 Application and results 


```python
data_copy = []
for model in algorithms:
    model.fit(messages_tfidf, y_train)
    pred_test = model.predict(test)
    data_copy.append(
        precision_recall_fscore_support(pred_test, y_test, average='macro'))

models_df = pd.DataFrame(data_copy, index=names)
models_df.columns = ['Precision', 'Recall', 'F_score', 'Support']
models_df
```

## 4.4 Supervised Learning Models


```python
#We normalize the features
X = normalize(df_featured.values[:, 2:17])
Y = list(df_featured.values[:, 17])

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

random_state = 43

names = [
    "Naive Bayesian Multinomial",
    "k-Nearest Neighbors",
    "Support Vector Machine",
    "Linear SVM",
    "RBF SVM",
    "Logistic Regression",
    "SGD Classifier",
]

algorithms = [
    MultinomialNB(),
    KNeighborsClassifier(n_neighbors=3),
    SVC(random_state=random_state),
    SVC(kernel="linear", random_state=random_state),
    SVC(kernel="rbf", random_state=random_state),
    LogisticRegression(random_state=random_state),
    SGDClassifier(max_iter=1000),
]

data_copy = []
i = 0
for model in algorithms:
    model.fit(X_train, y_train)
    pred_test = model.predict(X_test)
    print('The accuracy for', names[i], 'is',
          metrics.accuracy_score(pred_test, y_test))
    data_copy.append(
        precision_recall_fscore_support(pred_test, y_test, average='macro'))
    i = i + 1

models_df = pd.DataFrame(data_copy, index=names)
models_df.columns = ['Precision', 'Recall', 'F_score', 'Support']
models_df
```

As we can see, the models give very good accuracy which let us think that we came up with good models of prediction. It is a big trap because if we look at the predictions they are almost all zeros. This can also be seen through precision, recall and f_scores.

This is due to the fact that our model is very unbalanced. We were able to label only 18 people as POIs in 122 employees which represents a ratio of 14% only.

In the next parts, we will explore different techniques to find a solution to this problem.

***1st Idea : Over-Sampling***

Over-Sampling increases the number of instances in the minority class by replicating them in order to present a higher representation of the minority class in the sample. We duplicated 3 times the rows containing POIs in our training set.

Another idea would be to do under-sampling but since we have very few data we cannot afford to do so.

```python
random_state = 43

POIs = df_featured[df_featured['POI'] == 1]

df_featured_over = df_featured.append(POIs).append(POIs).append(POIs)

df_featured_over.reset_index(drop=True, inplace=True)

print(
    'New ratio of POIs after duplication is',
    len(df_featured_over[df_featured_over['POI'] == 1]) /
    len(df_featured_over))

#We normalize the features
X = normalize(df_featured_over.values[:, 2:17])
Y = list(df_featured_over.values[:, 17])

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)

names = [
    "Naive Bayesian Multinomial",
    "k-Nearest Neighbors",
    "Support Vector Machine",
    "Linear SVM",
    "RBF SVM",
    "Logistic Regression",
    "SGD Classifier",
]

algorithms = [
    MultinomialNB(),
    KNeighborsClassifier(n_neighbors=3),
    SVC(random_state=random_state),
    SVC(kernel="linear", random_state=random_state),
    SVC(kernel="rbf", random_state=random_state),
    LogisticRegression(random_state=random_state),
    SGDClassifier(max_iter=1000),
]

data_copy = []
i = 0
for model in algorithms:
    model.fit(X_train, y_train)
    pred_test = model.predict(X_test)
    data_copy.append(
        precision_recall_fscore_support(pred_test, y_test, average='macro'))
    i = i + 1

models_df = pd.DataFrame(data_copy, index=names)
models_df.columns = ['Precision', 'Recall', 'F_score', 'Support']
models_df
```

Even with duplicating the rows, we can see that almost all models give the same results which is quiete understandable, life would be so easy otherwise.

## 4.5 Unupervised Learning Models

For this part we decided not to use any prior knowlege of POIs. We therefore used KMEANS algorithm to perform the unsupervised task and try to classify people as corrupted or not. 

We will first try to use Kmeans using 2 features randomly selected, the stress score and the salary. We will then plot the results. We organized the centers and the labels in order to always have the same label names for each cluster when we run our code again.

```python
X = df_financial[["StressScore", "Salary"]]
array = X.values
DfTotalRows = len(array)
XPOI = df_financial["POI"]

#K means settings
kmeans = KMeans(n_clusters=2).fit(array)
index = np.argsort(kmeans.cluster_centers_.sum(axis=1))
labelsOrganized = np.zeros_like(index)
labelsOrganized[index] = np.arange(2)
kmeans.labels_ = labelsOrganized[kmeans.labels_]
CenterrsOranized = np.zeros_like(index)
CenterrsOranized[index] = np.arange(2)
kmeans.cluster_centers_ = kmeans.cluster_centers_[CenterrsOranized]

#Scatter using salary and stress score
centers = kmeans.cluster_centers_
plt.scatter(array[:, 0], array[:, 1], s=5)
plt.ylabel('Salary')
plt.xlabel('Stress Score')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=40, alpha=0.5)

#This score represents the number of correct guesses
score = 0

for i in range(DfTotalRows):
    y_kmeans = kmeans.predict([[array[i][0], array[i][1]]])
    if (y_kmeans == XPOI[i]):
        score += 1
        
#Results           
results = precision_recall_fscore_support(XPOI, kmeans.labels_)
result = (score / DfTotalRows) * 100
print(" **** ACCURACY is " + str(result) + "%****")
print(" **** PRECISION " + str(results[0][0] * 100) + "%****")
print(" **** RECALL " + str(results[1][0] * 100) + "%****")
print(" **** FBETA_SCORE " + str(results[2][0] * 100) + "%****")
```

We see a priori that there is a linear dependance between both features. We then applied K-Means on all the features to see if the results get better. 

We never expected to achieve such impreesive results.

```python
X = df_financial[[
    "Salary", "Bonus", "Long Term Incentive", "Deferred Income",
    "Deferral Payments", "Loan Advances", "Expenses", "Director Fees",
    "Exercised Stock options", "StressScore", "Restricted Stock",
    "Total Stock Value", "Nb_Emails_Sent",
    "Nb_Emails_Received", "Nb_Emails_FromPOI", "Nb_Emails_ToPOI", "StressScore"
]]

XPOI = df_financial["POI"]
array = X.values

#Kmeans setting
kmeans = KMeans(n_clusters=2).fit(array)
index = np.argsort(kmeans.cluster_centers_.sum(axis=1))
labelsOrganized = np.zeros_like(index)
labelsOrganized[index] = np.arange(2)
kmeans.labels_ = labelsOrganized[kmeans.labels_]
CenterrsOrganized = np.zeros_like(index)
CenterrsOrganized[index] = np.arange(2)
kmeans.cluster_centers_ = kmeans.cluster_centers_[CenterrsOrganized]

#This score represents the number of correct guesses
score = 0

#List of fraudulent employees 
fraudulent_persons=[]

#List of innoncent employees
non_fraudulent_persons=[]

for i in range(DfTotalRows):
    y_kmeans = kmeans.predict([[
        array[i][0], array[i][1], array[i][2], array[i][3], array[i][4],
        array[i][5], array[i][6], array[i][7], array[i][8], array[i][9],
        array[i][10], array[i][11], array[i][12], array[i][13], array[i][14],
        array[i][15], array[i][16]
    ]])  
    
    if ((y_kmeans[0]) == XPOI[i]):
        
        if(XPOI[i]==1):
            fraudulent_persons.append(df_financial['Name'][i])
        else :
            non_fraudulent_persons.append(df_financial['Name'][i])
        score += 1
        
#Results       
result = (score / DfTotalRows) * 100
print("**** ACCURACY is " + str(result) + "% *****")
results = precision_recall_fscore_support(XPOI, kmeans.labels_)
print("**** PRECISION " + str(results[0][0] * 100) + "% ****")
print("**** RECALL " + str(results[1][0] * 100) + "% ****")
print("**** FBETA_SCORE " + str(results[2][0] * 100) + "% ****")

print('\nList of employees predicted guilty',fraudulent_persons)
print('\nList of employees predicted innoncent',non_fraudulent_persons)
```
